```python
# Assignment week (4)four below are the questions and working for each questions
# Each [Question] is represented by(#)
#. Create function compute_average(score)
def compute_average(score):
  total = 0

  for score in scores:
    total = total + score

  average = total / len(scores)
  return average

scores = [67,70,89,77,55]
avg = compute_average(scores)
print("Average score:", avg)
```

```python
# create function check_pass_fail(sore)
def check_pass_fail(score):

  if score >= 60:
    return "pass"
  else:
    return "fail"

print(check_pass_fail(88))
print(check_pass_fail(50))
print(check_pass_fail(44))
print(check_pass_fail(60))
```

```python
# Apply to 5 students
scores = [67,70,89,77,55]

for score in scores:
  result = check_pass_fail(score)
  print("For a score of", score, "the result is", result)
```
